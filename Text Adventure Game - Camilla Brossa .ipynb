```python
 """
DocString:

1. Introduction and inspiration of the game:

This adventure game has two main inspirations: 
- Dante's masterpiece "The Divine Comedy", the imaginative journey 
of the poet through the afterlife, composed of Hell, Purgatory and Heaven;
- the NBC series "The Good Place", the story of Eleanor Shellstrop, 
who after her death finds herself in a highly selective Heaven-like utopia. 
There is only one problem: she doesn't belong there,
and you will step into her shoes to make the rightest decision.


2. Game structure:

The game develops in two different pathways: one toward Heaven and one toward Hell. 
The first scenario will determine which pathway you will be directed to.
The questions of each scenario all have a moral component.
Depending on your answer, you will be redirected to different scenarios,
and fictiously "doomed" to eternal salvation or damnation.
More realistically, you will understand if you're a good person or not.

****Different paths*****

3. Known bugs: none.
"""

import time                    #Importing the library 'time' to generate pauses during the game
from random import randint     #Importing the library 'random', and in particular 'randint' to generate random numbers for the last round

#Creating lists to define every acceptable user's answer throughout the game
guess_a = ['A','a']
guess_b = ['B','b']
guess_c = ['C', 'c']
guess_d = ['D', 'd']
guess_e = ['E', 'e']
guess1 = ['1', 'one', 'first', 'first one']
guess2 = ['2', 'two', 'second', 'second one']
guess3 = ['3', 'three', 'third', 'third one']

#Creating a list of lists, to include into a single list all possible valid user's answers
valid_answers = guess_a + guess_b + guess_c + guess_d + guess_e + guess1 + guess2 + guess3

#Intro - ask for user's information, explaining the scenario 
def intro():
    global user_name           #define these variables globally,
    global animal              #even outside of the function "intro"
    global place
    global food

    #The user name will be used in the game to make it more personalized
    #capitalize user's first name, in case he/she puts lower case
    user_name = input(prompt="""
    What's your name? 
    """)
    user_name = user_name.capitalize()        
    
    #using .lower() to put all of the following user inputs in lowcase since these variables are going to appear inside some text   
    #These inputs are going to be used to create dynamic strings and to make the game more entertaining
    animal = input(prompt="""
    What's the animal that scares you the most? 
    Please insert a singular name - i.e. snake: """)
    animal = animal.lower()                    
    
    place = input(prompt="""
    What's the last public place where you have been? - i.e. gym, supermarket, beach, etc.  
    """)
    place = place.lower()
    
    food = input(prompt="""
    What's the food you can't stand?
    """)
    food = food.lower()
    
    #Starting to print the first scenario
    print("‚ô¶Ô∏è"*30)
    print(" ")
    time.sleep(3)
    print("Imagine this:")
    
    time.sleep(3) #every time.sleep() function pauses the game for the number of seconds indicated in the brackets,
    #so that the user's experience improves by giving them small portions of text one at a time
    #Different lenghts of texts correspond to different lenghts of pauses
    
    print("""
    After a party with your friends, where you drank probably a bit too much üç∑, you wake up in someting that 
    looks like a waiting room, but completely white, with no walls or other people. 
    
    Just you, a chair and a door.
    """)
    time.sleep(5)
    print("""
    You decide to sit down, thinking that your body might need it after such a crazy night. But you suddenly realize
    that you don't have any hangover effects, not even a headache. It's like you don't feel your body anymore. 
    """)
    time.sleep(5)
    print("Someone opens the door and invites you to go in.")
    time.sleep(3)
    print("„Éª„Éª„Éª ")
    print("*"*80)
    time.sleep(3)
    
    #Using dynamic printing to include user's answers inside these following strings
    print(f"\n<< Hi, {user_name}. My name is Virgil and I am very happy to finally meet you. Welcome to Heaven!")
    time.sleep(2)
    print(" ")
    print(f"\nAround 43 minutes ago you died, while you were running away from a giant {animal} in a squalid {place}.")
    time.sleep(2)
    print(" ")
    print(f"\nNot sure why you were there, but the {food} you ate this morning caused you an allergic reaction and you fell from the stairs, and the {animal} ate you.")   
    time.sleep(3)
    print("""
    \nAnyways, I'm very happy to finally meet you. 
    During your time on Earth, you did so much good and you earned the label of a Good Person.
    You dedicated your life to help others. You spent 23 years in Syria helping orphans,
    you found a cure for AIDS, you even won the Nobel Peace Prize, twice.
    Here, you'll be happy for eternity. Enjoy. >> """)
    print("‚ô¶Ô∏è"*30)
    time.sleep(8)
    
    #Right after the introduction, the game will start automatically
    game_start()

#Defining the function to make the game start. This is the first scenario, that will determine which way
#the user is going to be directed to. If he/she chooses a, it will be the pathway to Heaven, but at each step
#the wrong answer can lead him to the other path, to Hell.
#Similarly, if during his/her path to Hell the user proves him/herself to improve, he/she can be redirected
#on the way to Heaven.
#Because everyone deserves a second possibility.

def game_start():
    print("""
    You're happy but extremely confused ü§î, because you know you didn't do any of the things Virgil mentioned.
    On Earth you were not a bad person, but you surely didn't win any Nobel Peace Prize.
    You then realize you don't deserve to be there, and you took someone else's spot.
    Someone that, because of this mistake, is now burning in hell.
    
    """)
    time.sleep(5)
    print("""
          \nWhat do you do?
    a. You decide to tell the truth and accept the consequences and your fate
    b. You don't say anything and you stay there, the other person is not your problem
    """)
    
    #The IF statement is used to make sure that the user's input is considered a valid answer,
    #and then part of the previously created list "valid_answers".
    #It is inside a WHILE LOOP, that will allow the user to insert again the input in case he/she enters
    #an invalid letter/word/input
    while True:   
        
        choice_1= input(prompt= "Enter 'a' or 'b': ")
        
        if choice_1 in valid_answers:  
            if choice_1 in guess_a:        #use of a NESTED CONDITIONAL to analize the possibilities 
                second_step()              #as long as the input is valid (pre-condition)
                break  

            #if the answer is a, this will lead the user to the second step of paradise path toward Heaven
            # and then the while loop is going to break
            
            elif choice_1 in guess_b:
                weird_things()
                break
            #if the answer is b, this will lead the user to a different scenario, the pathway to Hell, where weird things happen
            # and then the while loop is going to break
            

            #Use of else to detect bugs:
        else:
            print("Something went wrong in Game Start. If you entered an input different than 'a' or 'b', please try again.")
        #if the user inputs something invalid, the else will find the error and will allow the user to re-insert infinitely

        
#Defining the second step toward Heaven, in case the answer to the previous scenario was 'a'.
def second_step():
    
    #Use of dynamic printing to include user's inputs inside the text
    print(f"""
    Wise choice, {user_name}! üëèüèº
    
    Telling the truth was not easy, because you knew you could have been sent away from Heaven.
    But this is a first step toward paradise: you're actually earning your place here!
    However, your testing does not end here...
    """)
    time.sleep(5)
    
    print("""
    Now that you told the truth, you actually became a better person.
    But you also realize that all the bad things that you did during your time on Earth still have consequences
    today, and it is impossible to change them.
    """)
    time.sleep(5)            #a few ways to create suspence in order to improve user's experience
    print("""
    ...                          
    """)
    time.sleep(3)
    print("""Or maybe not!""")
    time.sleep(3)
    print("""
    You are now given the chance to go back to Earth and to your life.
    
    At one condition:
    You have to solve every unfinished business, fix each damage you created, apologize for every mistake you made.
    It's true, it will require a lot of time and effort, it won't be easy, but it will earn you a spot in Heaven.
    
    """)
    time.sleep(3)
    
    
    #Use of dynamic printing to include user's inputs inside the text, to make it more personalized
    
    print(f"""\nWhat do you choose, {user_name}?
    a. To go back to Earth and spend your time and effort trying to fix things, 
    and at the end of your life go deservedly to Heaven
    b. Go to the Purgatory, or the 'Middle Place', where you'll have a mediocre existence, 
    without problems nor happiness, just boredom, for the rest of time
    """)
    
    #The IF statement is used to make sure that the user's input is a valid answer,
    #and then part of the previously created list "valid_answers".
    #It is inside a WHILE LOOP, that will allow the user to insert again the input in case he/she enters
    #an invalid letter/word/input
    
    while True:
        
        choice_2= input(prompt= "Enter 'a' or 'b': ")

        if choice_2 in valid_answers:
            if choice_2 in guess_a:           #use of a NESTED CONDITIONAL to analize the possibilities 
                final_test()                  #as long as the input is valid (pre-condition)
                break

            #if the answer is a, this will lead him to the final step to go to Heaven and then
            #the loop will break

            
            elif choice_2 in guess_b:
                weird_things()
                break
                
            #if the answer is b, the user will be sent back to a scenario from the other path, the path to Hell,
            #and if he/she fails again will be sent to Hell 

            #Use of else to find bugs:
        else:
            print("Something went wrong. Make sure you inserted either 'a' or 'b'")
            #if the user inputs something invalid, the else will find the error and will allow the user to re-insert infinitely

            
#This scenario is the first step toward the pathway to Hell. If the user chose the wrong answer in the previous scenario,
#he/she has been redirected here. Weird things start happening, and the user still has the chance to prove
#himself/herself to be sent back to the pathway to Heaven
def weird_things():
    print(f"""
    \nWell, {user_name}, your decision was the easiest one to make. You chose to enjoy the pleasure
    of Heaven at the expenses of someone else who truly deserved this spot.
    You chose your own good, and that is okay.
    However, the perfect system of the afterlife understood that something is wrong, and weird things
    start to happen...
    """)
    time.sleep(5)
    print(f"""
    \nYou find yourself in the most horrific {place} you can possibly think of.
    When you look up, you see a gigantic {animal} with the head that looks like a {food},
    flying around and destroying everything.
    This certainly doesn't look like heaven!
    """)
    time.sleep(5)
    print(f"""
    You understand something is wrong, and before you can say a word, Virgil appears in front of you,
    with another quest...
    "These crazy things are happening because the Supreme Order understood you don't belong here.
    We must do something."
    """)
    time.sleep(5)
    print(f"""
    Again, you have two choices:
    you can decide to tell the truth, accept what the Supreme Order will decide for you,
    and put an end to this mess. Or you can ignore the numerous warnings and continue to omit the truth.
    Believe it or not, free will still exists in the afterlife, but every action has consequences.
    """)
    time.sleep(5)
    print("""
    What do you choose?
    a. Tell the truth
    b. Don't tell the truth
    """)
    
     #The IF statement is used to make sure that the user's input is a valid answer,
    #and then part of the previously created list "valid_answers".
    #It is inside a WHILE LOOP, that will allow the user to insert again the input in case he/she enters
    #an invalid letter/word/input
    
    while True:
        
        choice_3= input(prompt= "Enter 'a' or 'b': ")

        if choice_3 in valid_answers:
            if choice_3 in guess_a:           #use of a NESTED CONDITIONAL to analize the possibilities 
                second_step()                 #as long as the input is valid (pre-condition)
                break
                
            #If the choice is 'a', the right one, the user will be sent back to the pathway to Heaven to have a second
            #possibility and do better.
            #After that, the loop breaks
            

            elif choice_3 in guess_b:
                three_apples()
                break
                
            #If the choice is 'b', the user is keeping selecting the wrong answer, and its pathway
            #to Hell continues.
            #After that, the loop breaks.


            #Use of else to find bugs:
        else:
            print("Something went wrong. Make sure you inserted the letters 'a' or 'b'")
            #if the user inputs something invalid, the else will find the error and will allow the user to re-insert a valid input infinitely

            
#This is the last test to access Heaven, and it consists in a moral choice to prove
#the user's soul.
#Only one answer will lead him/her directly to Heaven ('e'); one will send him/her directly to Hell ('d');
#the othera will send him to an intermediate step of the path to Hell, where luck might help him/her (or not)
def final_test():
    print(f"""
    \nGood job, {user_name}! You proved to be a great person!
    Now, it's time for the very last test, to understand if you really deserve 
    a spot here, in Heaven.
    """)
    time.sleep(5)
    print("""
    Imagine this last scenario:
    
    For 24 hours you could do whatever you want - I really mean WHATEVER.
    There are not going to be consequences, you will get away with any action
    you do. Also, no one will ever know what you did, not even the Supreme Order.
    
    """)
    time.sleep(7)
    print("""
    What would you rather do?
    
    a. Getting wasted and driving as fast as you can
    b. Burning down the house of someone you hate
    c. Robbing a bank
    d. Kill someone you hate
    e. None of the above
    """)
    
    #The NESTED IF statement is used to make sure that the user's input is a valid answer,
    #and then part of the previously created list "valid_answers". 
    #Then, the following IFs, determine a scenario depending on the user's choice
    #It is inside a WHILE LOOP, that will allow the user to insert again the input in case he/she enters
    #an invalid letter/word/input
    
    while True:
        
        choice_4 = input(prompt= "Enter a letter: ")
        
        # Transform the input into lowercase, in order to use it in the IF statement
        choice_4 = choice_4.lower()

        if choice_4 in valid_answers:

            if choice_4 == 'a' or choice_4 =='b' or choice_4 == 'c':
                print("""
                Your answer to this last question was not exactly
                the best one. 

                However, before sending you to eternal damnation, I want to give you one last chance...
                    """) 

                time.sleep(3) # pauses for 3 seconds before executing the next function
                three_apples()
                break
               #If the user chooses one of the first 3 answers, he/she will be sent to the "Three Apples" step,
               #where he/she will have another chance before being sent to Hell.
               #Then the loop will break 

            elif choice_4 == 'd':
                hell()
                print("""
                You have been sent here because your choice during the last test made you
                unfit for Heaven. 
                Murder is never acceptable üôÖüèº.
                """)
                break
               #If the choice is 'd', murder, the user will be automatically sent to Hell, and an additional
               #text will be printed to let the user know that this was the worst possible result.
               #Then the loop will break

            elif choice_4 == 'e':
                heaven()
                print(f"""
                You reached the highest score possible in this game.
                You really are a great person, {user_name}!!

                Proud of you.
                """)
                break
               #If the choice is 'd', murder, the user will be automatically sent to Hell, and an additional
               #text will be printed to let the user know that this was the worst possible result.
               #Then the loop will break.

        else:
            print ("Something went wrong. Make sure you insert a letter between 'a' and 'e' ")
            #if the user inputs something invalid, the else will find the error 
            #and will allow the user to re-insert a valid input as many times as necessary, until he/she inserts
            #a valid input


#This scenario is the only one that only depends on luck.
#It lets the user choose a number and then confronts it with a random number generated by Python.
def three_apples():
    print(f"""
    \n{user_name}, this is you last chance to avoid burning in hell for the rest of Time.
    I gave you multiple opportunities to prove that you are a good person, and you failed.
    But I want to give you one last chance.
    This last step goes beyond every rational thought or decision. Follow your instinct.
    And may Fortune be with you.
    """)
    time.sleep(5)
    print("""
    You have three apples in front of you. 
    
    üçé üçé üçé
    
    Two of them are full of worms inside, only one is healthy and will guide you to Heaven.
    There is no way you can distinguish the good one from the bad ones.
    Choose one: first (1), second(2) or third(3)?
    """)
    
    #Generating the random number in the range 1 to 3, including the extremes.
    healthy_apple = randint(a=1,b=3)

    #Creating a WHILE loop to allow the user to re-insert an input multiple times in case it is invalid (not in the range)
    while True:
        
        #Transforming the user's input (automatically generated as a string) into an integer
        #in order to perform the comparison with the random number previously created
        choice_5 = int(input(prompt= "Enter a number from 1 to 3: "))
        
        #Making sure that only if the input is valid (in the range 1 to 3) the next steps follow
        if choice_5 >=1 and choice_5 <=3:

            if choice_5 == healthy_apple:
                heaven() 
                break
                
            #If the user's guess is equal to the randomly generated number, he/she will be sent to Heaven
            #then the loop will break

            elif choice_5 != healthy_apple:
                hell()
                break
            
            #If the user's guess is different from the randomly generated number, he/she will have terminated his/her
            #chances and will be sent to Hell.
            #Then the loop breaks


        else:
            print("Something went wrong. Make sure you inserted a number from 1 to 3.")
            
            #If the input is invalid the loop will continue to run infinitely, until the answer is in the range



        
#Win function, that happens if the user gets the higher score.
#The graphics have been taken from https://fsymbols.com/

def heaven():   
    time.sleep(3) #This will make the code wait for 3 seconds before printing, creating suspense
    print(f"""
    
    ‚ñë‚ñà‚ñà‚ïó‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïó‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñà‚ñà‚ïó‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ñà‚ïó‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ïó‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ÄÉ‚ÄÉ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë
    ‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïó‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ïî‚ïê‚ïê‚ïê‚ïê‚ïù‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ñà‚ñà‚ïë‚ñà‚ñà‚ïî‚ïê‚ïê‚ïê‚ïê‚ïù‚ÄÉ‚ÄÉ‚ïö‚ïê‚ïê‚ñà‚ñà‚ïî‚ïê‚ïê‚ïù‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó
    ‚ñë‚ïö‚ñà‚ñà‚ïó‚ñà‚ñà‚ñà‚ñà‚ïó‚ñà‚ñà‚ïî‚ïù‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ïö‚ïê‚ïù‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ïî‚ñà‚ñà‚ñà‚ñà‚ïî‚ñà‚ñà‚ïë‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñë‚ÄÉ‚ÄÉ‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë
    ‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ïî‚ïê‚ñà‚ñà‚ñà‚ñà‚ïë‚ñë‚ñà‚ñà‚ïî‚ïê‚ïê‚ïù‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïó‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë‚ïö‚ñà‚ñà‚ïî‚ïù‚ñà‚ñà‚ïë‚ñà‚ñà‚ïî‚ïê‚ïê‚ïù‚ñë‚ñë‚ÄÉ‚ÄÉ‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë
    ‚ñë‚ñë‚ïö‚ñà‚ñà‚ïî‚ïù‚ñë‚ïö‚ñà‚ñà‚ïî‚ïù‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ïö‚ñà‚ñà‚ñà‚ñà‚ñà‚ïî‚ïù‚ïö‚ñà‚ñà‚ñà‚ñà‚ñà‚ïî‚ïù‚ñà‚ñà‚ïë‚ñë‚ïö‚ïê‚ïù‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ÄÉ‚ÄÉ‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ïö‚ñà‚ñà‚ñà‚ñà‚ñà‚ïî‚ïù
    ‚ñë‚ñë‚ñë‚ïö‚ïê‚ïù‚ñë‚ñë‚ñë‚ïö‚ïê‚ïù‚ñë‚ñë‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù‚ñë‚ïö‚ïê‚ïê‚ïê‚ïê‚ïù‚ñë‚ñë‚ïö‚ïê‚ïê‚ïê‚ïê‚ïù‚ñë‚ïö‚ïê‚ïù‚ñë‚ñë‚ñë‚ñë‚ñë‚ïö‚ïê‚ïù‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù‚ÄÉ‚ÄÉ‚ñë‚ñë‚ñë‚ïö‚ïê‚ïù‚ñë‚ñë‚ñë‚ñë‚ïö‚ïê‚ïê‚ïê‚ïê‚ïù‚ñë

    ‚ñà‚ñà‚ïó‚ñë‚ñë‚ñà‚ñà‚ïó‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ïó‚ñë‚ñë‚ñë‚ñà‚ñà‚ïó‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñà‚ñà‚ñà‚ïó‚ñë‚ñë‚ñà‚ñà‚ïó
    ‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ïî‚ïê‚ïê‚ïê‚ïê‚ïù‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ïî‚ïê‚ïê‚ïê‚ïê‚ïù‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ïë
    ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïë‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïë‚ïö‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ïî‚ïù‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñë‚ñà‚ñà‚ïî‚ñà‚ñà‚ïó‚ñà‚ñà‚ïë
    ‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïë‚ñà‚ñà‚ïî‚ïê‚ïê‚ïù‚ñë‚ñë‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïë‚ñë‚ïö‚ñà‚ñà‚ñà‚ñà‚ïî‚ïù‚ñë‚ñà‚ñà‚ïî‚ïê‚ïê‚ïù‚ñë‚ñë‚ñà‚ñà‚ïë‚ïö‚ñà‚ñà‚ñà‚ñà‚ïë
    ‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ïö‚ñà‚ñà‚ïî‚ïù‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñà‚ñà‚ïë‚ñë‚ïö‚ñà‚ñà‚ñà‚ïë
    ‚ïö‚ïê‚ïù‚ñë‚ñë‚ïö‚ïê‚ïù‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù‚ïö‚ïê‚ïù‚ñë‚ñë‚ïö‚ïê‚ïù‚ñë‚ñë‚ñë‚ïö‚ïê‚ïù‚ñë‚ñë‚ñë‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù‚ïö‚ïê‚ïù‚ñë‚ñë‚ïö‚ïê‚ïê‚ïù, 
    
    {user_name}! üòá
    
    ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ
    ‚ú®üåà    ‚ú®üåà   ‚ú®üåà   ‚ú®üåà   ‚ú®üåà   ‚ú®üåà   ‚ú®üåà   ‚ú®üåà   ‚ú®üåà    ‚ú®üåà    ‚ú®üåà    ‚ú®üåà    ‚ú®üåà
    ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ‚ú∑‚úØ
                        
                    ________________¬∂¬∂¬∂¬∂_________¬∂¬∂¬∂¬∂
                    ______________¬∂¬∂1111¬∂¬∂_____¬∂¬∂1111¬∂¬∂
                    _____________¬∂11111111¬∂___¬∂11111111¬∂
                    ____________¬∂1111111111¬∂¬∂¬∂1111111111¬∂
                    ___________¬∂111111111¬∂¬∂¬∂¬∂¬∂¬∂¬∂111111111¬∂
                    ___________¬∂111111¬∂¬∂¬∂1111111¬∂¬∂¬∂111111¬∂
                    __________¬∂11111¬∂¬∂1111111111111¬∂¬∂1111¬∂
                    __________¬∂1111¬∂11111111111111111¬∂111¬∂
                    ______¬∂___¬∂111¬∂1111¬∂¬∂11111¬∂¬∂¬∂¬∂1111¬∂11¬∂___¬∂
                    _____¬∂____¬∂11¬∂111¬∂¬∂__¬∂111¬∂___¬∂¬∂1111¬∂1¬∂____¬∂
                    _____¬∂___¬∂11¬∂111¬∂____¬∂¬∂111¬∂¬∂¬∂__¬∂111¬∂11¬∂___¬∂
                    _____¬∂¬∂__¬∂11¬∂11¬∂_______¬∂¬∂¬∂¬∂¬∂____¬∂11¬∂11¬∂__¬∂¬∂
                    _____¬∂_¬∂¬∂11¬∂111¬∂_________________¬∂111¬∂11¬∂¬∂_¬∂
                    ______¬∂1111¬∂11¬∂___________________¬∂11¬∂1111¬∂
                    _______¬∂¬∂¬∂1¬∂1¬∂_____________________¬∂1¬∂1¬∂¬∂¬∂
                    _________¬∂111¬∂__¬∂¬∂¬∂___________¬∂¬∂¬∂__¬∂111¬∂
                    ________¬∂11¬∂1¬∂_¬∂___¬∂_________¬∂___¬∂_¬∂1¬∂11¬∂
                    ________¬∂1¬∂¬∂1¬∂__¬∂¬∂¬∂___________¬∂¬∂¬∂__¬∂1¬∂¬∂1¬∂
                    ________¬∂11¬∂1¬∂_¬∂_¬∂¬∂¬∂_________¬∂¬∂¬∂_¬∂_¬∂1¬∂11¬∂
                    ________¬∂11¬∂1¬∂_¬∂¬∂¬∂¬∂¬∂_________¬∂¬∂¬∂¬∂¬∂_¬∂1¬∂11¬∂
                    _________¬∂111¬∂¬∂_¬∂¬∂¬∂__¬∂__¬∂__¬∂__¬∂¬∂¬∂_¬∂¬∂¬∂11¬∂
                    ¬∂_________¬∂111¬∂_______¬∂¬∂_¬∂¬∂_______¬∂111¬∂_________¬∂
                    ¬∂¬∂_________¬∂¬∂¬∂¬∂¬∂____¬∂_______¬∂____¬∂¬∂¬∂¬∂¬∂_________¬∂¬∂
                    ¬∂_¬∂_____________¬∂____¬∂_____¬∂____¬∂_____________¬∂_¬∂
                    ¬∂__¬∂¬∂_____¬∂¬∂¬∂¬∂¬∂__¬∂____¬∂¬∂¬∂¬∂¬∂____¬∂__¬∂¬∂¬∂¬∂¬∂_____¬∂¬∂__¬∂
                    ¬∂____¬∂¬∂¬∂¬∂¬∂_____¬∂__¬∂¬∂_________¬∂¬∂__¬∂_____¬∂¬∂¬∂¬∂¬∂____¬∂
                    ¬∂_______________¬∂___¬∂¬∂¬∂¬∂¬∂¬∂¬∂¬∂¬∂___¬∂_______________¬∂
                    _¬∂_______________¬∂¬∂¬∂__¬∂___¬∂__¬∂¬∂¬∂_______________¬∂
                    ¬∂_¬∂¬∂___¬∂________¬∂¬∂_____¬∂¬∂¬∂_____¬∂¬∂________¬∂___¬∂¬∂_¬∂
                    ¬∂___¬∂¬∂¬∂________¬∂1¬∂_____________¬∂1¬∂________¬∂¬∂¬∂___¬∂
                    ¬∂______________¬∂11¬∂___________¬∂11¬∂______¬∂¬∂______¬∂
                    _¬∂_____¬∂_______¬∂111¬∂¬∂_______¬∂¬∂111¬∂_______¬∂_____¬∂
                    __¬∂¬∂¬∂¬∂¬∂________¬∂11111¬∂¬∂¬∂¬∂¬∂¬∂¬∂11111¬∂________¬∂¬∂¬∂¬∂¬∂
                    __¬∂____________¬∂11¬∂11111111111¬∂11¬∂___________¬∂
                    __¬∂___________¬∂11¬∂1111111111111¬∂11¬∂___________¬∂
                    ___¬∂__________¬∂11¬∂1111111111111¬∂11¬∂__________¬∂
                    ____¬∂¬∂_____¬∂¬∂¬∂111¬∂1111111111111¬∂111¬∂¬∂¬∂_____¬∂¬∂
                    ______¬∂¬∂¬∂¬∂¬∂__¬∂111¬∂1111111111111¬∂111¬∂__¬∂¬∂¬∂¬∂¬∂
                    ____________¬∂111¬∂111111111111111¬∂111¬∂
                    __________¬∂¬∂1111¬∂111111111111111¬∂1111¬∂¬∂
                    _________¬∂¬∂_¬∂11¬∂11111111111111111¬∂11¬∂_¬∂¬∂
                    ________¬∂1¬∂__¬∂1¬∂11111111111111111¬∂1¬∂__¬∂1¬∂
                    _______¬∂111¬∂¬∂¬∂¬∂1111111111111111111¬∂¬∂¬∂¬∂111¬∂
                    ______¬∂1111111¬∂1111111111111111111¬∂1111111¬∂
                    ______¬∂111111¬∂111111111111111111111¬∂111111¬∂
                    ______¬∂11111¬∂11111111111111111111111¬∂11111¬∂
                    _______¬∂111111111111111111111111111111111¬∂
                    ________¬∂1111111111111111111111111111111¬∂
                    _________¬∂¬∂111111111111111111111111111¬∂¬∂
                    ___________¬∂¬∂¬∂111111111111111111111¬∂¬∂¬∂
                    ______________¬∂¬∂¬∂¬∂1111111111111¬∂¬∂¬∂¬∂
                    __________________¬∂¬∂¬∂¬∂¬∂¬∂¬∂¬∂¬∂¬∂¬∂¬∂¬∂



    
    """)
    time.sleep(6)
    while True:
        replay = input(prompt = """Would you like to play again? Insert 'yes' or 'no':\n""")
        replay = replay.lower()
        if replay == 'yes':
            intro()
            break
        elif replay == 'no':
            print("""Thank you for playing! See you soon.""")
            break
        else:
            print("Invalid answer. Please retry.")

#Fail function, that happens if the user fails all the stages 
#The graphics have been taken from https://fsymbols.com/

def hell():
    time.sleep(3) #This will make the code wait for 3 seconds before printing, creating suspense
    print(f"""
    
    ‚ñà‚ñà‚ïó‚ñë‚ñë‚ñë‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ïó‚ñë‚ñë‚ñë‚ñà‚ñà‚ïó‚ÄÉ‚ÄÉ‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ÄÉ‚ÄÉ‚ñà‚ñà‚ïó‚ñà‚ñà‚ñà‚ïó‚ñë‚ñë‚ñà‚ñà‚ïó‚ÄÉ‚ÄÉ‚ñà‚ñà‚ïó‚ñë‚ñë‚ñà‚ñà‚ïó‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñà‚ñà‚ïó‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïó‚ñë‚ñë‚ñë‚ñë‚ñë
    ‚ïö‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ïî‚ïù‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ÄÉ‚ÄÉ‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó‚ñà‚ñà‚ïî‚ïê‚ïê‚ïê‚ïê‚ïù‚ÄÉ‚ÄÉ‚ñà‚ñà‚ïë‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñà‚ñà‚ïë‚ÄÉ‚ÄÉ‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ïî‚ïê‚ïê‚ïê‚ïê‚ïù‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñë‚ñë
    ‚ñë‚ïö‚ñà‚ñà‚ñà‚ñà‚ïî‚ïù‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ÄÉ‚ÄÉ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïî‚ïù‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñë‚ÄÉ‚ÄÉ‚ñà‚ñà‚ïë‚ñà‚ñà‚ïî‚ñà‚ñà‚ïó‚ñà‚ñà‚ïë‚ÄÉ‚ÄÉ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïë‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñë‚ñë
    ‚ñë‚ñë‚ïö‚ñà‚ñà‚ïî‚ïù‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ÄÉ‚ÄÉ‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïë‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïó‚ñà‚ñà‚ïî‚ïê‚ïê‚ïù‚ñë‚ñë‚ÄÉ‚ÄÉ‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë‚ïö‚ñà‚ñà‚ñà‚ñà‚ïë‚ÄÉ‚ÄÉ‚ñà‚ñà‚ïî‚ïê‚ïê‚ñà‚ñà‚ïë‚ñà‚ñà‚ïî‚ïê‚ïê‚ïù‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ñë‚ñë
    ‚ñë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñë‚ïö‚ñà‚ñà‚ñà‚ñà‚ñà‚ïî‚ïù‚ïö‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïî‚ïù‚ÄÉ‚ÄÉ‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ÄÉ‚ÄÉ‚ñà‚ñà‚ïë‚ñà‚ñà‚ïë‚ñë‚ïö‚ñà‚ñà‚ñà‚ïë‚ÄÉ‚ÄÉ‚ñà‚ñà‚ïë‚ñë‚ñë‚ñà‚ñà‚ïë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ïó
    ‚ñë‚ñë‚ñë‚ïö‚ïê‚ïù‚ñë‚ñë‚ñë‚ñë‚ïö‚ïê‚ïê‚ïê‚ïê‚ïù‚ñë‚ñë‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù‚ñë‚ÄÉ‚ÄÉ‚ïö‚ïê‚ïù‚ñë‚ñë‚ïö‚ïê‚ïù‚ïö‚ïê‚ïù‚ñë‚ñë‚ïö‚ïê‚ïù‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù‚ÄÉ‚ÄÉ‚ïö‚ïê‚ïù‚ïö‚ïê‚ïù‚ñë‚ñë‚ïö‚ïê‚ïê‚ïù‚ÄÉ‚ÄÉ‚ïö‚ïê‚ïù‚ñë‚ñë‚ïö‚ïê‚ïù‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù‚ïö‚ïê‚ïê‚ïê‚ïê‚ïê‚ïê‚ïù
    {user_name}!
    
    Enjoy eternal damnation! üòà
    
    
    By the way, for the Contrappasso Law (afterlife retaliation), 
    you will be eating {food}, while being bitten by a gigantic {animal} for 
    the rest of time. 
    
   
                                ‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë
                                ‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë
                                ‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë
                                ‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë
                                ‚ñë‚ñë‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñë‚ñë
                                ‚ñë‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñë
                                ‚ñà‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà
                                ‚ñà‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà
                                ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà
                                ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà
                                ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà
                                ‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñë
                                ‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñë‚ñà‚ñà‚ñà‚ñà‚ñë‚ñë
                                ‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñÄ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñÄ‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë
                                ‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚îÄ‚îÄ‚ñÄ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñÄ‚îÄ‚îÄ‚ñà‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë
                                ‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚îÄ‚îÄ‚îÄ‚ñà‚ñà‚ñà‚ñà‚ñà‚îÄ‚îÄ‚îÄ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë
                                ‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñÑ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñÑ‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë
                                ‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë
                                ‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë
                                ‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë
                                ‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë
                                ‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë
                                ‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚îÄ‚îÄ‚ñÄ‚ñÄ‚ñÄ‚îÄ‚îÄ‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë
                                ‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚îÄ‚ñà‚ñà‚ñà‚ñà‚ñà‚îÄ‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë
                                ‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚îÄ‚ñà‚ñà‚ñà‚îÄ‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë
                                ‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñà‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë
                                ‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë‚ñë



    """)
    while True:
        replay = input(prompt = """Would you like to play again? Insert 'yes' or 'no':\n""")
        replay = replay.lower()
        if replay == 'yes':
            intro()
            break
        elif replay == 'no':
            print("""Thank you for playing! See you soon.""")
            break
        else:
            print("Invalid answer. Please retry.")
    

####################GAME STARTS##########################

intro()



    
```
